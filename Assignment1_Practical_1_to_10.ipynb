```python
'''
Practical 2:
Write a program to check whether a triangle is right angled triangle or not using pythagorus theorem
'''

a = float(input("Enter the length of the first side: "))
b = float(input("Enter the length of the second side: "))
c = float(input("Enter the length of the third side: "))

if a > b and a > c:
    hypot = a;
    side1 = b;
    side2 = c;

elif b > a and b > c:
    hypot = b;
    side1 = a;
    side2 = c;

else:
    hypot = c;
    side1 = b;
    side2 = a;

if hypot ** 2 == side1 ** 2 + side2 ** 2:
    print("This triangle is right-angled.")
else:
    print("This triangle is not right-angled.")
```

```python
'''
Practical 4:
Write a python program to check given number is a prime number or not
Use while loop and if-else
Show factors that are not prime
'''

num = int(input("Enter the number to be checked: "))
counter = 2
count = 0

while counter < num:
    if num % counter == 0:
        count += 1
        print(f"{num} is divisible by {counter}")
    counter += 1

if num > 1:
    if count > 0:
        print(f"{num} is NOT a Prime Number.")
    else:
        print(f"{num} is a Prime Number.")
else:
    print(f"{num} is NOT a Prime Number.")
```

```python
'''
Practical 6:
Write a python program to construct pascal's triangle for a given number of rows. Use for loop
Output with rows: 5
'''

'''
Pascal's Triangle Example (with 5 rows):
                1
            1       1
        1       2       1
    1       3       3       1
1       4       6       4       1
'''

rows = int(input("Please enter the number of rows: "))

for i in range(rows):
    for j in range(rows, i, -1):
        print("  ", end="")
    num = 1

    for j in range(i + 1):
        print(num, end="   ")
        num = num * (i - j) // (j + 1)
    print("\n",end="")
```

```python
'''
Practical 7: Data structures in python
B. Given a list of Integers , find all pairs whose sum is equal to a given target
'''

numList = []
targetList = []

while True:
    num = input("Please enter a number or leave blank to proceed: ")
    if not num:
        break
    numList.append(int(num))

target = int(input("\nEnter the target number: "))

for num in numList:
    for numY in numList:
        if num + numY == target:
            targetList.append([num, numY])

for element in targetList:
    for elementY in targetList:
        if element == elementY or element.reverse() == elementY:
            targetList.remove(elementY)

print(f"\nTarget List: {targetList}")
```

```python
'''
Practical 7: Data structures in python
C. Rotate elements in a tuple by 'n' positions to the left and right
'''

inList = []
while True:
    num = input("Please enter a number or press enter to proceed: ")
    if not num:
        break
    inList.append(int(num))

tupleX = tuple(inList)

change = int(input("\nEnter the change: \n"))
change = change % len(tupleX)
tupleX_left = tupleX[change:] + tupleX[:change]
tupleX_right = tupleX[-change:] + tupleX[:-change]

print(f"Original Tuple: {tupleX}")
print(f"Left-Shift Tuple: {tupleX_left}")
print(f"Right-Shift Tuple: {tupleX_right}")
```

```python
'''
Practical 8
Write a python program that generates power set of a given list of Integers *recursively*
a. Keep only the groups where adding all numbers gives an even number
b. Arrange the groups so that the smallest groups comes last.
'''

numList = []

while True:
    num = input("Please enter a number or press enter to proceed: ")
    if not num:
        break
    numList.append(int(num))
print(f"List of numbers: {numList}")

def powerSet(numList):
    if not numList:
        return [[]]
    first = numList[0]
    rest = numList[1:]
    powerSet1 = powerSet(rest)
    subSets = [[first] + subset for subset in powerSet1]
    return powerSet1 + subSets

def evenSet(power_set):
    return [subset for subset in power_set if sum(subset) % 2 == 0]

def arrangeSet(listX):
    return sorted(listX, key=len, reverse=True)

def finalSet(inputList):
    power = powerSet(inputList)
    even = evenSet(power)
    arranged = arrangeSet(even)
    return arranged

result = finalSet(numList)
print("\nFinal Result:")
for group in result:
    print(f"{group}", end="\t")
```

```python
'''
Practical 9:
Write a python program to implement a simple shopping cart that allows users to:
add items
view the cart and
calculate the total bill
using recursion and lambda functions.
'''

def viewCart():
    print("\n| Product\t\t\t | Quantity\n-------------------------------------------------------")
    for product in shoppingCart:
        print(f"| {product}\t | {shoppingCart[product]}")

def addToCart(product, quantity):
    if product not in productDict.keys():
        print("Product not in the catalogue.")
        return 1
    if product not in shoppingCart.keys():
        shoppingCart[product] = quantity
    else:
        shoppingCart[product] += quantity

def totalPrice(cart, index=0):
    if index == len(cart):
        return 0
    subtotal = cart[index][1] * cart[index][2]
    return subtotal + totalPrice(cart, index + 1)

applyDiscount = lambda total: total * 0.9 if total > 100000 else total
productDict = {"M1 MACBOOK AIR": 55000, "M2 MACBOOK AIR": 70000, "M3 MACBOOK AIR": 80000, "M4 MACBOOK AIR": 100000}

userName = input("Please enter your name: ").lower().capitalize()
print(f"Hi {userName}, Welcome to Aayush's macStore!")

while True:
    startQ = input("Would you like to start shopping? (yes or no): ")

    if startQ.lower() == "no":
        print("We are sorry to hear that!")
        break

    elif startQ.lower() == "yes":
        shoppingCart = {}
        while True:
            opt = int(
                input("Choose an option to continue:\n1. View Cart\n2. Add item(s)\n3. Caculate total\n4. Checkout"))

            if opt == 1:
                viewCart()

            elif opt == 2:
                print("| Product\t\t\t | Price\n-------------------------------------------------------")
                for item in productDict:
                    print(f"| {item}\t | {productDict[item]}")
                item = input("Please enter the item to add from the catalogue: ")
                qty = int(input("Please enter the quantity of the item: "))
                addToCart(item.upper(), qty)

            elif opt == 3:
                if not shoppingCart:
                    print("The cart is empty.")
                else:
                    cart_items = [(product, productDict[product], quantity) for product, quantity in
                                  shoppingCart.items()]
                    total = totalPrice(cart_items)
                    finalPrice = applyDiscount(total)
                    print(f"Final Total: Rs. {finalPrice}")

            elif opt == 4:
                del shoppingCart
                print("Thank you for shopping with us!")
                break

            else:
                print("Please enter a valid option.")
        break
    else:
        print("Invalid input. Please say yes or no.")
```

```python
'''
Practical 10:
Write a python program to print
a solid diamond shape
and a hollow diamond shape
for a given number of rows
'''

rows = int(input("Enter the no. of rows: "))

for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")
    for j in range(i * 2 - 1):
        print("*", end="")
    print("\n", end="")

for i in range(rows - 1, 0, -1):
    for j in range(rows - i):
        print(" ", end="")
    for j in range(i * 2 - 1):
        print("*", end="")
    print("\n", end="")
rows = int(input("Enter the no. of rows: "))

for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(1, i * 2):
        if k == i * 2 - 1 or k == 1:
            print("*", end="")
        else:
            print(" ", end="")
    print("\n", end="")

for i in range(rows, 1, -1):
    for j in range(rows - i):
        print(" ", end="")
    for k in range(i * 2 - 1):
        if k == i * 2 - 3 or k == 1:
            print("*", end="")
        else:
            print(" ", end="")
    print("\n", end="")
```
