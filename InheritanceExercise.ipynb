```python
# Hybrid

class Test1:
    def test1_method(self):
        print("Blah")

class Test2 (Test1):
    def test2_method(self):
        print("Blah2")

class Test3 ():
    def test3_method(self):
        print("Blah3")

class Test4 (Test2, Test3):
    def test4_method(self):
        print("Blah4")

obj = Test4()
obj.test4_method()
obj.test3_method()
obj.test2_method()
obj.test1_method
```
