```python
def viewCart():
    print("\n| Product\t\t\t | Quantity\n-------------------------------------------------------")
    for product in shoppingCart:
        print(f"| {product}\t | {shoppingCart[product]}")
```

```python
def addToCart(product, quantity):
    if product not in productDict.keys():
        print("Product not in the catalogue.")
        return 1
    if product not in shoppingCart.keys():
        shoppingCart[product] = quantity
    else:
        shoppingCart[product] += quantity
```

```python
def totalPrice(cart, index=0):
    if index == len(cart):
        return 0
    subtotal = cart[index][1] * cart[index][2]
    return subtotal + totalPrice(cart, index + 1)

applyDiscount = lambda total: total * 0.9 if total > 100000 else total
```

```python
productDict = {"M1 MACBOOK AIR": 55000, "M2 MACBOOK AIR": 70000, "M3 MACBOOK AIR": 80000, "M4 MACBOOK AIR": 100000}

userName = input("Please enter your name: ").lower().capitalize()
print(f"Hi {userName}, Welcome to Aayush's macStore!")

while True:
    startQ = input("Would you like to start shopping? (yes or no): ")

    if startQ.lower() == "no":
        print("We are sorry to hear that!")
        break

    elif startQ.lower() == "yes":
        shoppingCart = {}
        while True:
            opt = int(input("Choose an option to continue:\n1. View Cart\n2. Add item(s)\n3. Caculate total\n4. Checkout"))

            if opt == 1:
                viewCart()

            elif opt == 2:
                print("| Product\t\t\t | Price\n-------------------------------------------------------")
                for item in productDict:
                    print(f"| {item}\t | {productDict[item]}")
                item = input("Please enter the item to add from the catalogue: ")
                qty = int(input("Please enter the quantity of the item: "))
                addToCart(item.upper(), qty)

            elif opt == 3:
                if not shoppingCart:
                    print("The cart is empty.")
                else:
                    cart_items = [(product, productDict[product], quantity) for product, quantity in shoppingCart.items()]
                    total = totalPrice(cart_items)
                    finalPrice = applyDiscount(total)
                    print(f"Final Total: Rs. {finalPrice}")

            elif opt == 4:
                del shoppingCart
                print("Thank you for shopping with us!")
                break

            else:
                print("Please enter a valid option.")
        break
    else:
        print("Invalid input. Please say yes or no.")
```
