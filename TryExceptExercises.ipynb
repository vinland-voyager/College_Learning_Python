```python
# Write a python program handle index error and key error getting input from user
```

```python
dictX = {"NAME" : "Aayush", "AGE" : 21}
listX = [1, 2, 3]
try:
    testX = input("Enter the key to check: ")
    testX = testX.upper()
    testY = int(input("Enter the index to print: "))
    print(dictX[testX])
    print(listX[testY])
except KeyError:
    print("Key not found.")
except IndexError:
    print("Index invalid.")
except:
    print("Something went wrong.")
finally:
    print("End of program.")
```

```python
# Write a python function that takes two numbers as input and returns the result of the division, handle the ZeroDivisionError exception by returning a message "cannot divide by zero", if the denominator is zero.
```

```python
numerator = int(input("Enter the numerator: "))
denominator = int(input("Enter the denominator: "))

def ZeroDivisionExercise(numerator, denominator):
    try:
        quotient = numerator / denominator
        print("The result is: ", quotient)
    except ZeroDivisionError:
        print("Cannot divide by zero.")
    except:
        print("Something went wrong.")
    finally:
        print("Program end.")

ZeroDivisionExercise(numerator, denominator)
```
