```python
# Write a program to create an array using numpy to get 5 numbers from user, store in an array and find product of all numbers as stored in array
```

```python
import numpy as np
X = np.array([])
for i in range(5):
    num = int(input("Enter a number: "))
    X = np.append(X, num)
print(X)
product = 1
for i in range(5):
    product *= X[i]
print(product)
```

```python
# Write a python to create an array, getting 10 integers from user on runtime. Display all the integer with their square+5
```

```python
Y = np.array([])
for i in range(10):
    num = int(input("Enter a number: "))
    Y = np.append(Y, num)
for i in range(10):
    print((Y[i]**2)+5)
```
