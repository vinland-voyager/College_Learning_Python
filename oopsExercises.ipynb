```python
'''
Create a python class named circle with attributes radius and color.
Implement methods area() to calculate the area of the circle
and display() to display the radius and color of the circle
'''
```

```python
class Circle:
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color

    def area(self):
        import math
        return math.pi * self.radius * self.radius

radius = float(input("Enter the radius: "))
color = input("Enter the color: ")
test = Circle(radius, color)
test.area()
```
